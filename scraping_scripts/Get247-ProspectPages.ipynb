```python
from bs4 import BeautifulSoup
import requests
import lxml
import time
import json
import os.path
from os import path
```

```python
#This goes thru a full folder as defined by the confFolder variable
headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'}
confFolder = ['bigten', 'bigtwelve']
```

```python
#Starting by just getting the team page
#But you are going to have to go another two layers down
for conf in confFolder:
    playerDir = "..//html//247//" + conf + "//recruits//"
    for file in os.listdir(playerDir):
        print(file)
        gameSoup = BeautifulSoup(open(playerDir + file, "r", encoding='utf8').read(), 'lxml')
        starter = "as-a-prospect"
        fileName = file.split('_')
        playerName = fileName[0]
        team = fileName[1]
        year = fileName[3].split('.')[0]
        thisFile = "..//html//247//" + conf + "//prospects//" + playerName + "_" + team + "_secondhop_" + year + ".html"
        if (path.isfile(thisFile)):
            print("file exists")
        else:
            print("fetching new file")
            if (gameSoup.find("a", class_="view-profile-link") is not None):
                prospectLink = gameSoup.find("a", class_="view-profile-link")['href']
                #print(prospectLink)
                req = requests.get(prospectLink, headers=headers)
                with open("..//html//247//" + conf + "//prospects//" + playerName + "_" + team + "_secondhop_" + year + ".html", "w") as write_file:
                    write_file.write(req.text)
                time.sleep(3)                
```
