```python
from bs4 import BeautifulSoup
import requests
import lxml
import time
import json
import os.path
from os import path
import core_constants as cc
```

```python
years = cc.get_defYears()
headers= cc.get_header()
conf = 'acc'

schoolsList = cc.get_schoolsList()
playerDir = cc.get_htmlDir('247', conf, 'recruits')
```

```python
#Starting by just getting the team page
#But you are going to have to go another two layers down

for file in os.listdir(playerDir):
    print(file)
    gameSoup = BeautifulSoup(open(playerDir + file, "r", encoding='utf8').read(), 'lxml')
    starter = "as-a-prospect"
    fileName = file.split('_')
    playerName = fileName[0]
    team = fileName[1]
    year = fileName[3].split('.')[0]
    thisFile = cc.get_htmlDir('247', conf, 'prospects') + playerName + "_" + team + "_secondhop_" + year + ".html"
    if (path.isfile(thisFile)):
        print("file exists")
    else:
        if (gameSoup.find("a", class_="view-profile-link") is not None):
            prospectLink = gameSoup.find("a", class_="view-profile-link")['href']
            #print(prospectLink)
            req = requests.get(prospectLink, headers=headers)
            cc.save_files(thisFile, req.text)
            time.sleep(3)                
```
