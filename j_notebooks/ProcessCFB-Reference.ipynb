```python
from bs4 import BeautifulSoup
import requests
import lxml
import time
import json
import os.path
from os import path
```

```python
playerIndexDir = "..//html//cfb_ref//playerIndex//"
```

```python
cfb_players = []
for file in os.listdir(playerIndexDir):
    
    playerSoup = BeautifulSoup(open(playerIndexDir + file, "r").read(), 'lxml')
    playerList = playerSoup.find("div", id="div_players")
    for player in playerList.find_all("p"):
        cfb_player = {}
        date = ((player.text[-12:]).strip()).split("-")[0][1:]
        if (int(date) > 2000):
            cfb_player['year'] = date
            for playerInfo in player.find_all("a"):
                if ("players" in playerInfo['href']):          
                    cfb_player['playerName'] = playerInfo.text.strip()
                    cfb_player['player_url'] = playerInfo['href']
                elif ("schools" in playerInfo['href']):
                    cfb_player['school'] = playerInfo.text.strip()
            cfb_players.append(cfb_player)
```

```python
with open("..//scraped_json//cfb_ref_list.json", "w") as write_file:
    json.dump(cfb_players, write_file)
```

```python
def requestPage (url):
    baseURL = 'https://www.sports-reference.com/'
    headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'}
    r = requests.get(baseURL + url, headers=headers)
    request = {}
    request['status_code'] = r.status_code
    request['text'] = r.text
    return request
```
