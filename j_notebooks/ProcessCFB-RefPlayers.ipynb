```python
from bs4 import BeautifulSoup
import requests
import lxml
import time
import json
import os.path
from os import path
import re
```

```python
#available conferences: bigtwelve, bigten, sec
conference = 'bigten'
confPlayersDir = '..//html//cfb_ref//playerPages//' + conference + '//'
cfbPlayers = []
for file in os.listdir(confPlayersDir):
    cfbPlayer = {}
    playerSoup = BeautifulSoup(open(confPlayersDir + file, "r", encoding='windows-1252', errors="ignore").read(), 'lxml')
    playerInfo = playerSoup.find('div', itemtype="https://schema.org/Person")
    cfbPlayer['playerName'] = (playerSoup.find('h1', itemprop="name").text).strip()
    print(cfbPlayer['playerName'])
    for info in (playerInfo.find_all("p")):
        #if (info.find("strong").text == 'Schools'):
        infoList = []
        height = re.compile(r"[0-9]+-[0-9]{2}")
        for item in info.stripped_strings:
            item = item.replace("\n", "")
            item = item.replace(" ", "")
            infoList.append(item)
        if (infoList):
            if (infoList[0] == 'School' or infoList[0] == 'Schools'):
                if (len(infoList) > 3):
                    cfbPlayer['school'] = [infoList[2], infoList[4]]
                else:
                    cfbPlayer['school'] = [infoList[2]]
            if (infoList[0] == 'Position'):
                cfbPlayer['position'] = infoList[1][1:]
            if (re.findall(height, infoList[0])):
                height = infoList[0].split("-")
                height = ((int(height[0]) * 12) + int(height[1]))
                cfbPlayer['height'] = height
                cfbPlayer['weight'] = infoList[2][:-2]
                
    playerStats = playerSoup.find('table', class_="stats_table")
    gamesPlayed = 0
    if (playerStats is not None):
        for row in playerStats.find_all('td', {'data-stat':'g'}):
            if (row.text == ''):
                gamesPlayed = 0 + gamesPlayed
            else:
                gamesPlayed = int(row.text) + gamesPlayed

        cfbPlayer['gamesPlayed'] = gamesPlayed
    cfbPlayers.append(cfbPlayer)

#print(cfbPlayers)
    #for player in playerList.find_all("p"):
```

```python
with open("..//scraped_json//cfb_ref_" + conference + ".json", "w", encoding="utf-8") as write_file:
    write_file.write(json.dumps(cfbPlayers))
```
